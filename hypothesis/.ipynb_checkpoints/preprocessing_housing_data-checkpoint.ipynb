```python
import os
import matplotlib as mpl
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from matplotlib.ticker import FuncFormatter
```

```python
path = os.getcwd()
parent_path = os.path.dirname(path)
spaces = "\n\n"

# Read Prices Dataset
# `read_parquet()` requires `parquet` package to be installed.
df = pd.read_parquet(f"{parent_path}/datasets/housing_prices.parquet")
```

```python
print("--- Columns and types: ")
print(df.dtypes, end="\n\n")

print("--- Unique areas and their frequencies: ")
print(df["area"].value_counts(), end=spaces)

print("--- Unique regions and their frequencies: ")
print(df["region"].value_counts(), end=spaces)
```

```python
df.isnull().sum()
```

```python
# Starting the cleaning process

columns_to_keep = [
    "date",
    "year_build",
    "purchase_price",
    "city",
    "region"
]

# Keep only instances between 2014 and 2024
start_date = "2014-01-01"
end_date = "2024-12-31"

# Keep only Zealand region
target_region = "Zealand"

df_clean = df[
    (df["date"] >= start_date) &
    (df["date"] <= end_date) &
    (df["region"] == target_region)
][columns_to_keep].copy()

# Sort and reset index
df_clean = df_clean.sort_values(by="date").reset_index(drop=True)

# Modify `date` column to keep only the year.
df_clean["date"] = df_clean["date"].dt.year
```

```python
# We need to flip the above structure in order to make it easier for pandas to work with
# when using the .map() function.
city_to_municipality = {}
for muni, cities in municipality_groups.items():
    for city in cities:
        city_to_municipality[city] = muni

df_clean["municipality"] = df_clean["city"].map(city_to_municipality)
```

### SOME PLOTS

```python
mpl.rcParams.update({
    "text.usetex": True,
    "font.family": "serif",
    "font.size": 12,
    "axes.titlesize": 16,
    "axes.labelsize": 12,
    "pgf.rcfonts": False,
    "pgf.texsystem": "pdflatex",
    "text.latex.preamble": r"\usepackage{amsmath}",
})

yearly_prices = df_clean.groupby("date")["purchase_price"].median()

fig, ax = plt.subplots(figsize=(10, 6))
sns.lineplot(data=yearly_prices, linewidth=2.5, ax=ax)

ax.set_title(r"\textbf{Median Housing Price Trend (2014--2024)}")
ax.set_ylabel(r"Price (DKK)")
ax.set_xlabel(r"Year")
ax.grid(True, linestyle="--", alpha=0.7)

def millions(x, pos):
    return r"%1.1fM" % (x * 1e-6)

ax.yaxis.set_major_formatter(FuncFormatter(millions))

fig.tight_layout()

fig.savefig("housing_trend.pgf")
plt.show()
```

```python
sorted_order = (
    df_clean.groupby("municipality")["purchase_price"]
    .median()
    .sort_values(ascending=False)
    .index
)

fig, ax = plt.subplots(figsize=(14, 8))

sns.boxplot(
    data=df_clean,
    x="municipality",
    y="purchase_price",
    order=sorted_order,
    ax=ax,
)

ax.set_title(r"\textbf{Price Distribution by Municipality}")
ax.set_ylabel(r"Purchase Price (DKK)")
ax.set_xlabel("")

ax.tick_params(axis="x", rotation=45)
ax.grid(True, axis="y", linestyle="--", alpha=0.5)

fig.tight_layout()
fig.savefig("price_distribution_by_municipality.pgf")
plt.show()
```

```python
order = df_clean["municipality"].value_counts().index

fig, ax = plt.subplots(figsize=(12, 6))

sns.countplot(
    data=df_clean,
    x="municipality",
    order=order,
    ax=ax,
)

ax.set_title(r"\textbf{Number of Sales per Municipality}")
ax.set_ylabel(r"Number of Transactions")
ax.set_xlabel("")

ax.tick_params(axis="x", rotation=45)
ax.grid(True, axis="y", linestyle="--", alpha=0.5)

fig.tight_layout()
fig.savefig("number_of_sales_per_municipality.pgf")
plt.show()
```
